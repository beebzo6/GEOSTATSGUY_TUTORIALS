```python
import tensorflow as tf
import numpy as np

# Define X and y for this cell to ensure they are available
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])
X = tf.constant(X, dtype=tf.float32)
y = tf.constant(y, dtype=tf.float32)

model=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(100,activation='relu'),
        tf.keras.layers.Dense(100,activation='relu'),
        tf.keras.layers.Dense(100,activation='relu'),
        tf.keras.layers.Dense(1)
    ])
model.compile(loss="mae",
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
# Reshape X and y to be 2-dimensional (batch_size, num_features)
X_reshaped = tf.expand_dims(X, axis=-1)
y_reshaped = tf.expand_dims(y, axis=-1)

model.fit(X_reshaped, y_reshaped, epochs=100)
```

```python

def mae (y_pr,y):
  total_sum=0
  total_sample=len(y_pr)
  summation=0
  for i in total_sample:
     ss=y[i] -y_pr[i]
     total_sum+=abs(ss)
  summation=total_sum/total_sample
  return summation

```
